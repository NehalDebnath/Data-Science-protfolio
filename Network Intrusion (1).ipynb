# Project Name: Network Intrusion
    The main aim of the project is to inspect all the network activity identify suspicious patterns that may indicate a network or system attack.

```python
# importing libraries
import pandas as pd
pd.set_option("display.max_columns", None)  # command to show all the variables\columns
```

```python
# reading data
ni = pd.read_csv("C:/DATA SCIENCE/Python/PyCharm Community Edition 2019.3.2/MY/Network_Intrusion.csv")
ni.shape
```

```python
ni.head()
```

```python
ni["class"].value_counts()
```

```python
#Renaming The Coloumn Class as Label
ni.rename(columns={"class": "label"}, inplace=True)
ni["label"].value_counts()
```

```python
#Checking For Null Values
ni.isnull().sum()  # no null values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
ni.select_dtypes(include=['object']).columns
```

```python
ni[ni.select_dtypes(include=['object']).columns] = ni[ni.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
ni.head()
```

```python
ni.shape
```

```python
ni_x = ni.iloc[:, 0:41]
ni_x
```

```python
ni_y = ni.iloc[:, 41]
ni_y
```

```python
from sklearn.model_selection import train_test_split
ni_x_train, ni_x_test, ni_y_train, ni_y_test = train_test_split(ni_x, ni_y, test_size=.2)
```

```python
##model building
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(ni_x_train, ni_y_train)
```

```python
pred_values = log_reg.predict(ni_x_test)
pred_values
```

```python
from sklearn.metrics import confusion_matrix, classification_report

tab_net = confusion_matrix(pred_values, ni_y_test)
tab_net
```

```python
tab_net.diagonal().sum() / tab_net.sum() * 100
```

```python
#Checking Accuracy
print(classification_report(pred_values, ni_y_test))
```

```python
log_reg.coef_
```

```python
log_reg.coef_.shape
```

```python
variables_coef = pd.DataFrame({"Variables": ni_x_train.columns, "Values": list(log_reg.coef_)[0]})
variables_coef = variables_coef.sort_values("Values", ascending=False)
variables_coef
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(pred_values, ni_y_test)
```

```python
pred_prob = log_reg.predict_proba(ni_x_test)
pred_prob = pd.DataFrame(pred_prob)  # to read this add the 1st and the 2ndno of the same line u will get the ans as 1
# probablity of falling in class zero 0.07004301, probablity of falling in class one 0.92995699
```

```python
from sklearn.metrics import roc_curve
fpr , tpr , thre = roc_curve(ni_y_test , pred_prob.iloc[: , 1])
import matplotlib.pyplot as plt
```

```python
plt.plot(fpr,tpr)
plt.grid()
plt.xlabel("FPR Values", size = 15)
plt.ylabel("TPR Values",size = 15)
plt.title("AUROC Curve")
plt.text(x = .4 , y = .5 , s ="Area Under Curve Value is .88" )
```

```python
##Building Decision Tree

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=5)
dt.fit(ni_x_train,ni_y_train)
```

```python
pred_dt = dt.predict(ni_x_test)
mat1 = confusion_matrix(pred_dt, ni_y_test)
mat1
```

```python
mat1.diagonal().sum() / mat1.sum() *100
```

```python
dt.feature_importances_
```

```python
var_imp = pd.DataFrame({"Variable" : ni_x_train.columns,"Score" : dt.feature_importances_})
var_imp.sort_values("Score",ascending=False)
```
